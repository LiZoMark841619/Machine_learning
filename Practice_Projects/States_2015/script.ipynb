```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('states_2015.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
columns = df.columns.to_list()[1:2] + df.columns.to_list()[9:]
```

```python
sns.heatmap(df[columns].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
plt.clf()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(df[columns], diag_kind='kde', plot_kws = {'alpha': 0.7, 's': 30, 'edgecolor': 'k'})
plt.show()
plt.clf()
```

```python
for column in columns:
    sns.histplot(df[column], color='darkblue', bins=12)
    plt.title(f'Distribution of "{column}" in {len(df)} countries in 2015')
    plt.vlines(np.mean(df[column]), color='red', label='Mean', ymax=20, ymin=0)
    plt.vlines(np.median(df[column]), color='blue', label='Median',ymax=20, ymin=0)
    plt.xlabel('Ranges')
    plt.ylabel('Count [nums]')
    plt.legend()
    plt.show()
    plt.clf()
```
